```python
%matplotlib inline
import pandas as pd
import os
import csv
pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)

train = pd.read_csv(os.path.join(os.getcwd(),'train_set.csv'), delim_whitespace=False)
test = pd.read_csv(os.path.join(os.getcwd(),'test_set.csv'), delim_whitespace=False)
```

```python
train
```

```python
col = train.drop(["Id","PRICE"],axis=1).columns
```

```python
#Train data内での重複データ数
train.duplicated(subset=col,keep=False).sum()
```

```python
#Test data内での重複データ数
test.duplicated(subset=col,keep=False).sum()
```

```python
#重複データを削除
train_dup = train[~train.duplicated(subset=col)]
train_dup.duplicated(subset=col,keep=False).sum()
```

```python
#重複データを削除
test_dup = test[~test.duplicated(subset=col)]
test_dup.duplicated(subset=col,keep=False).sum()
```

```python
#データの結合
df_all = pd.concat([train_dup,test_dup])
```

```python
######Data leakage######
#Train-Test間で352行の重複あり
#つまり176行のデータがTrain-Testで同条件
#PRICEも同額なのでは？
df_all.duplicated(subset=col,keep=False).sum()
```

```python
#重複している行の抽出
df_leak = df_all[df_all.duplicated(subset=col,keep=False)]
df_train = df_leak[df_leak["Id"] < 49108]
df_test = df_leak[df_leak["Id"] >= 49108]
```
